```python
import pandas as pd
```

```python
import re
```

```python
df = pd.read_csv("videohype.csv")
pd.set_option('display.max_rows', None)
```

```python
df['Copy'] = df['Copy'].str.lower()
```

```python
df['Copy'] = df['Copy'].str.replace(r'[^\w\s]+', '')
```

```python
#Word Count
topwords = pd.Series(' '.join(df.Copy).split()).value_counts()
topwords
```

```python
#number of letters in each decsription
longest = [df['Copy'].astype(str).str.len().sort_values(ascending=False).head(10)]
#longest
```

```python
#Longest or shortest character count
lettercount = df['Copy'].astype(str).str.len()
titles = df['Video']
newset = pd.merge(titles, lettercount, left_index=True, right_index=True)
newset.sort_values(by='Copy', ascending=False).head(10)
```

```python
#word count by video
wordcount = df['Copy'].str.split().str.len()
newset2 = pd.merge(titles, wordcount, left_index=True, right_index=True)
newset2.columns=['Video', 'Wordcount']
newset2.sort_values(by='Wordcount', ascending=False).head(10)
```
